```python
# Cargamos paquetes
import numpy as np
```

### Calcular producto punto

El primer paso es calcular el logit, *h*, a partir del producto punto. La fórmula explícita es la siguiente:

$$ h = W X +b $$

```python
# TODO: Implementa la función h usando numpy!

def combinacion_lineal (W,X , b):
    h = np.dot(X, W) + b
    return h
```

### Función de activación

Para este ejemplo utilizaremos la función sigmoide como función de activación.

$$ f(x) =  \frac{\mathrm{1} }{\mathrm{1} + e^{-x} }  $$ 

```python
# TODO: Implementar la función de activación. 
# En este ejemplo utilizaremos la función sigmoide.
def sigmoide(h):
    sg = 1/(1 + np.e**(-h))
    return sg
```

### Neurona simple

Implementación de la neurona simple

```python
def neurona(W, X, b, activacion):
    h = combinacion_lineal(W, X, b)
    return activacion(h)
```

## Probar la red

Ahora definamos unos pesos y veamos el resultado de una pasada frontal (fordward pass).

```python
# Definamos unos pesos y sesgo
inputs = np.array([0.7, -0.3])
weights = np.array([0.1, 0.8])
bias = -0.1

# TODO: Pasada frontal
# define la función de activación a usar 
# define la neurona
# calcula la salida
output = neurona(weights, inputs, bias, sigmoide)

print('Output:')
print(output)
```

## Pruebas extra
- Define una función de activación tangente hiperbólica. 
- Contesta: ¿La tanh y sigmoide producen las mismas salidas?

```python
# TODO: Completar el código
def tanh(h):
    tanh = (np.e**(h) - np.e**(-h))/(np.e**(h) + np.e**(-h))
    return tanh
# TODO: Realiza de nuevo la inferencia con la nueva función de activación
```

```python
output = neurona(weights, inputs, bias, tanh)
print('Output:')
print(output)
```
